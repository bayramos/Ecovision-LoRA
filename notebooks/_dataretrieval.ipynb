```python
# Download Sample Spacenet_2 Data

import os
import requests
import zipfile

# Create data folders
os.makedirs("data/spacenet", exist_ok=True)

url = "https://spacenet-dataset.s3.amazonaws.com/spacenet/SN2_buildings/train/AOI_2_Vegas"
out_path = "../data/spacenet/AOI_2_Vegas"

# Download if not already downloaded
if not os.path.exists(out_path):
    print("Downloading SpaceNet sample...")
    r = requests.get(url, stream=True)
    with open(out_path, "wb") as f:
        for chunk in r.iter_content(chunk_size=8192):
            f.write(chunk)
else:
    print("AOI_2_Vegas.zip already exists.")

# Extract
with zipfile.ZipFile(out_path, 'r') as z:
    z.extractall("data/spacenet/")

print("Extraction complete.")
```

```python
# Download Sample Spacenet_4 Data  

import os
import requests
import tarfile

# Create folders
os.makedirs("../data/spacenet/AOI_4_Shanghai", exist_ok=True)

url = "https://spacenet-dataset.s3.amazonaws.com/spacenet/SN2_buildings/tarballs/SN2_buildings_train_AOI_4_Shanghai.tar.gz"
out_path = "../data/spacenet/AOI_4_Shanghai.tar.gz"

# Download file
if not os.path.exists(out_path):
    print("Downloading AOI_4_Shanghai...")
    r = requests.get(url, stream=True)
    with open(out_path, "wb") as f:
        for chunk in r.iter_content(chunk_size=8192):
            f.write(chunk)
else:
    print("File already exists:", out_path)

# Extract TAR.GZ
print("Extracting...")
with tarfile.open(out_path, "r:gz") as tar:
    tar.extractall("../data/spacenet/AOI_4_Shanghai")

print("Shanghai AOI extraction complete.")
```
